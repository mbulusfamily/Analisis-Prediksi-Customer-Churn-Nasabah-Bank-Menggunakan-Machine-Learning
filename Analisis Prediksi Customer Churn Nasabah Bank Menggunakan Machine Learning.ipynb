Penelitian ini mengembangkan sistem prediksi churn nasabah menggunakan machine learning dengan dataset Bank Churners yang terdiri dari 10,127 nasabah dengan 16.1% churn rate.

**Metodologi** penelitian mencakup preprocessing anti-leakage, feature engineering, dan perbandingan tiga algoritma (Logistic Regression, Random Forest, XGBoost) menggunakan cross-validation dengan metrik Precision-Recall AUC. Sistem juga mengimplementasikan optimasi threshold berbasis ROI untuk memaksimalkan business value.

**Hasil** menunjukkan bahwa model XGBoost memberikan performa terbaik dengan ROC AUC 0.993 dan PR AUC 0.967. Optimasi ROI menghasilkan threshold 0.990 dengan ROI 8.84x, precision 98%, dan recall 58%. Analysis feature importance mengidentifikasi faktor utama churn: jumlah transaksi (18.9%), saldo bergulir (10.3%), dan nilai transaksi (7.6%).

**Kontribusi** penelitian ini adalah sistem prediksi churn yang production-ready dengan interpretability tinggi, framework optimasi ROI, dan perbandingan pendekatan binary vs. churn-only modeling. Sistem ini dapat meningkatkan efektivitas program retensi nasabah dengan targeting yang presisi dan cost-effective.

````python
# Global Setup
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.metrics import (confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score, average_precision_score, roc_curve, precision_recall_curve)
import joblib
import scipy.stats as stats
try:
    import shap
    SHAP_AVAILABLE = True
except ImportError:
    SHAP_AVAILABLE = False
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
warnings.filterwarnings('ignore', category=UserWarning)
warnings.filterwarnings('ignore', category=FutureWarning)
# Create output folders if not exist
for folder in ['models', 'reports', 'artifacts']:
    os.makedirs(folder, exist_ok=True)
````

````python
# 1) Load Data & Validasi Target (anti-leakage)
print("🔄 Loading and validating data...")

# Try to load data with fallback
try:
    df = pd.read_csv(r"c:\Users\ASUS\Downloads\BankChurners.csv\BankChurners.csv")
    data_source = "BankChurners.csv"
except FileNotFoundError:
    try:
        df = pd.read_csv("data_ecommerce_customer_churn.csv")
        data_source = "data_ecommerce_customer_churn.csv"
    except FileNotFoundError:
        print("❌ Neither BankChurners.csv nor data_ecommerce_customer_churn.csv found!")
        raise

print(f"✅ Loaded data from: {data_source}")
print(f"📊 Original shape: {df.shape}")

# Handle target variable
if 'Attrition_Flag' in df.columns:
    print("🎯 Found Attrition_Flag column - mapping to Churn")
    df['Churn'] = df['Attrition_Flag'].map({
        "Attrited Customer": 1,
        "Existing Customer": 0
    }).astype(int)
    print(f"   Mapping completed: {df['Churn'].value_counts().to_dict()}")
elif 'Churn' in df.columns:
    print("🎯 Found existing Churn column")
    df['Churn'] = df['Churn'].astype(int)
else:
    print("❌ No target column found (Attrition_Flag or Churn)")
    raise ValueError("Target column not found")

# Drop leakage columns if they exist
leakage_cols = [
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'
]

dropped_leakage = []
for col in leakage_cols:
    if col in df.columns:
        df.drop(col, axis=1, inplace=True)
        dropped_leakage.append(col)

if dropped_leakage:
    print(f"🗑️  Dropped leakage columns: {len(dropped_leakage)} columns")
else:
    print("✅ No leakage columns found to drop")

# Filter loyal customers (≥6 months)
tenure_col = None
if 'Months_on_book' in df.columns:
    tenure_col = 'Months_on_book'
elif 'tenure_months' in df.columns:
    tenure_col = 'tenure_months'

if tenure_col:
    original_len = len(df)
    df = df[df[tenure_col] >= 6].copy()
    print(f"🔍 Filtered loyal customers (≥6 months): {original_len} → {len(df)} rows")
else:
    print("⚠️  No tenure column found (Months_on_book/tenure_months) - skipping loyalty filter")

# Drop unique ID columns
id_cols = ['CLIENTNUM', 'CustomerID']
dropped_ids = []
for col in id_cols:
    if col in df.columns:
        df.drop(col, axis=1, inplace=True)
        dropped_ids.append(col)

if dropped_ids:
    print(f"🗑️  Dropped ID columns: {dropped_ids}")

# Also drop Attrition_Flag if it exists (we already have Churn)
if 'Attrition_Flag' in df.columns:
    df.drop('Attrition_Flag', axis=1, inplace=True)
    print("🗑️  Dropped original Attrition_Flag column")

print(f"\n📈 Final dataset shape: {df.shape}")
print(f"📊 First 5 rows:")
display(df.head())

# Churn proportion
churn_counts = df['Churn'].value_counts()
churn_prop = df['Churn'].mean()
print(f"\n🎯 Churn distribution:")
print(f"   Non-churn (0): {churn_counts[0]} ({(1-churn_prop)*100:.1f}%)")
print(f"   Churn (1): {churn_counts[1]} ({churn_prop*100:.1f}%)")

# Column types
numeric_features = df.select_dtypes(include=[np.number]).columns.tolist()
if 'Churn' in numeric_features:
    numeric_features.remove('Churn')

categorical_features = df.select_dtypes(include=['object', 'category', 'bool']).columns.tolist()

print(f"\n📋 Feature types:")
print(f"   Numeric features ({len(numeric_features)}): {numeric_features}")
print(f"   Categorical features ({len(categorical_features)}): {categorical_features}")
````

````python
# 2) Split Data yang Benar (hindari leakage)
print("🔀 Splitting data with anti-leakage practices...")

# Separate features and target
X = df.drop('Churn', axis=1)
y = df['Churn']

print(f"📊 Features shape: {X.shape}")
print(f"🎯 Target shape: {y.shape}")

# Stratified train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2, 
    stratify=y, 
    random_state=RANDOM_STATE
)

print(f"\n✅ Train-test split completed:")
print(f"   Train: {X_train.shape[0]} samples")
print(f"   Test: {X_test.shape[0]} samples")
print(f"   Train churn rate: {y_train.mean():.3f}")
print(f"   Test churn rate: {y_test.mean():.3f}")

print("\n⚠️  IMPORTANT: No preprocessing applied yet - all will be done in Pipeline to prevent leakage!")
````

````python
# 3) Preprocessing & Imbalance
print("⚙️  Setting up preprocessing pipeline...")

# Auto-detect column types from training data
categorical_features = X_train.select_dtypes(include=['object', 'category', 'bool']).columns.tolist()
numeric_features = X_train.select_dtypes(include=[np.number]).columns.tolist()

print(f"📊 Detected feature types:")
print(f"   Numeric features ({len(numeric_features)}): {numeric_features[:5]}{'...' if len(numeric_features) > 5 else ''}")
print(f"   Categorical features ({len(categorical_features)}): {categorical_features}")

# Create preprocessing pipelines
numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Combine preprocessors
preprocessor = ColumnTransformer([
    ('num', numeric_pipeline, numeric_features),
    ('cat', categorical_pipeline, categorical_features)
])

print(f"✅ Preprocessor created with {len(numeric_features)} numeric and {len(categorical_features)} categorical features")

# Calculate class imbalance from training data only
pos = (y_train == 1).sum()
neg = (y_train == 0).sum()
scale_pos_weight = float(neg / pos)

print(f"\n⚖️  Class imbalance in training data:")
print(f"   Negative class (0): {neg} samples")
print(f"   Positive class (1): {pos} samples")
print(f"   Scale pos weight: {scale_pos_weight:.3f}")

# Note: We'll get feature names after fitting the preprocessor in the pipeline
````

````python
# 4) Model Candidates + Tuning (CV pakai PR AUC)
print("🤖 Setting up model candidates and hyperparameter tuning...")

# Create model candidates
models = {
    'LogisticRegression': LogisticRegression(
        solver='saga', 
        max_iter=5000, 
        class_weight='balanced',
        random_state=RANDOM_STATE
    ),
    'RandomForest': RandomForestClassifier(
        random_state=RANDOM_STATE,
        n_jobs=-1
    ),
    'XGBoost': XGBClassifier(
        objective='binary:logistic',
        eval_metric='logloss',
        n_jobs=-1,
        scale_pos_weight=scale_pos_weight,
        random_state=RANDOM_STATE
    )
}

# Create pipelines
pipelines = {}
for name, model in models.items():
    pipelines[name] = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])

print(f"✅ Created {len(pipelines)} pipelines: {list(pipelines.keys())}")

# Hyperparameter search spaces
param_grids = {
    'LogisticRegression': {
        'classifier__C': stats.loguniform(1e-3, 1e2),
        'classifier__penalty': ['l1', 'l2']
    },
    'RandomForest': {
        'classifier__n_estimators': [300, 500, 800],
        'classifier__max_depth': [None, 10, 20, 30],
        'classifier__min_samples_split': [2, 5, 10],
        'classifier__min_samples_leaf': [1, 2, 4],
        'classifier__max_features': ['sqrt', 'log2']
    },
    'XGBoost': {
        'classifier__n_estimators': [400, 600, 800],
        'classifier__max_depth': [3, 4, 5, 6],
        'classifier__learning_rate': [0.01, 0.05, 0.1, 0.2],
        'classifier__subsample': [0.7, 0.8, 1.0],
        'classifier__colsample_bytree': [0.7, 0.9, 1.0],
        'classifier__gamma': [0, 1, 5],
        'classifier__reg_lambda': [1, 3, 5]
    }
}

# Cross-validation setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

print(f"\n🔍 Starting hyperparameter tuning with {cv.n_splits}-fold CV...")
print("📊 Primary scoring: average_precision (PR AUC)")

# Perform tuning for each model
tuning_results = {}
best_models = {}

for name in pipelines.keys():
    print(f"\n🔄 Tuning {name}...")
    
    # RandomizedSearchCV
    search = RandomizedSearchCV(
        pipelines[name],
        param_grids[name],
        n_iter=20,  # Reduced for faster execution
        cv=cv,
        scoring='average_precision',
        n_jobs=-1,
        random_state=RANDOM_STATE,
        verbose=0
    )
    
    # Fit
    search.fit(X_train, y_train)
    
    # Store results
    best_models[name] = search.best_estimator_
    tuning_results[name] = {
        'best_score': search.best_score_,
        'best_params': search.best_params_,
        'cv_results': search.cv_results_
    }
    
    print(f"   ✅ Best CV PR AUC: {search.best_score_:.4f}")

# Compare models and select best
print(f"\n📊 Model Comparison (CV PR AUC):")
print("-" * 40)
best_model_name = None
best_score = -1

for name, results in tuning_results.items():
    score = results['best_score']
    print(f"{name:15s}: {score:.4f}")
    if score > best_score:
        best_score = score
        best_model_name = name

print(f"\n🏆 Best model: {best_model_name} (PR AUC: {best_score:.4f})")
best_pipeline = best_models[best_model_name]
````

````python
# 5) Fit Ulang & Evaluasi Holdout
print("🔄 Refitting best model and evaluating on holdout test set...")

# Refit best pipeline on full training data (already done in RandomizedSearchCV)
# Get predictions on test set
y_proba = best_pipeline.predict_proba(X_test)[:, 1]
y_pred = best_pipeline.predict(X_test)

print(f"✅ Generated predictions on {len(X_test)} test samples")

# Evaluate at default threshold (0.5)
print(f"\n📊 Evaluation at default threshold (0.5):")
print("-" * 45)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()

print(f"Confusion Matrix:")
print(f"   TN: {tn:4d}  FP: {fp:4d}")
print(f"   FN: {fn:4d}  TP: {tp:4d}")

# Metrics
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_proba)
pr_auc = average_precision_score(y_test, y_proba)

print(f"\nMetrics:")
print(f"   Precision: {precision:.4f}")
print(f"   Recall:    {recall:.4f}")
print(f"   F1-Score:  {f1:.4f}")
print(f"   ROC AUC:   {roc_auc:.4f}")
print(f"   PR AUC:    {pr_auc:.4f}")

# Plot ROC and PR curves
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_proba)
ax1.plot(fpr, tpr, linewidth=2, label=f'ROC AUC = {roc_auc:.3f}')
ax1.plot([0, 1], [0, 1], 'k--', alpha=0.5)
ax1.set_xlabel('False Positive Rate')
ax1.set_ylabel('True Positive Rate')
ax1.set_title('ROC Curve')
ax1.legend()
ax1.grid(True, alpha=0.3)

# PR Curve
precision_curve, recall_curve, _ = precision_recall_curve(y_test, y_proba)
ax2.plot(recall_curve, precision_curve, linewidth=2, label=f'PR AUC = {pr_auc:.3f}')
ax2.axhline(y=y_test.mean(), color='k', linestyle='--', alpha=0.5, label='Baseline')
ax2.set_xlabel('Recall')
ax2.set_ylabel('Precision')
ax2.set_title('Precision-Recall Curve')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('reports/roc_pr_curves.png', dpi=100, bbox_inches='tight')
plt.show()
plt.close()

# Save metrics
import json
metrics_default = {
    'model': best_model_name,
    'threshold': 0.5,
    'confusion_matrix': {'tn': int(tn), 'fp': int(fp), 'fn': int(fn), 'tp': int(tp)},
    'precision': float(precision),
    'recall': float(recall),
    'f1_score': float(f1),
    'roc_auc': float(roc_auc),
    'pr_auc': float(pr_auc)
}

with open('artifacts/metrics_default.json', 'w') as f:
    json.dump(metrics_default, f, indent=2)

print(f"\n💾 Saved default metrics to artifacts/metrics_default.json")
````

````python
# 6) Tuning Threshold Berbasis ROI (Wajib)
print("💰 Tuning threshold based on business ROI...")

# Business parameters - CONFIGURABLE VALUES
RETAINED_VALUE = 500_000  # Value saved per retained customer (Rp)
OFFER_COST = 50_000      # Cost per targeted customer (Rp)

# You can modify these values based on your business case:
retained_value = RETAINED_VALUE
offer_cost = OFFER_COST

# Business functions
def business_gain(tp, fp, retained_value, offer_cost):
    """Calculate business gain"""
    revenue = tp * retained_value
    cost = (tp + fp) * offer_cost
    return revenue - cost

def roi(tp, fp, retained_value, offer_cost):
    """Calculate ROI"""
    cost = (tp + fp) * offer_cost
    return business_gain(tp, fp, retained_value, offer_cost) / max(cost, 1e-9)

print(f"💼 Business assumptions:")
print(f"   Retained value per customer: Rp {retained_value:,}")
print(f"   Campaign cost per customer:  Rp {offer_cost:,}")

# Threshold sweep
print(f"\n🔍 Sweeping thresholds from 0.01 to 0.99...")

threshold_results = []
best_roi_info = {'threshold': None, 'roi': -1e9}
high_recall_threshold = None

for threshold in np.linspace(0.01, 0.99, 99):
    # Get predictions at this threshold
    y_pred_thr = (y_proba >= threshold).astype(int)
    
    # Confusion matrix
    tp = ((y_test == 1) & (y_pred_thr == 1)).sum()
    fp = ((y_test == 0) & (y_pred_thr == 1)).sum()
    fn = ((y_test == 1) & (y_pred_thr == 0)).sum()
    tn = ((y_test == 0) & (y_pred_thr == 0)).sum()
    
    # Metrics
    prec = tp / max(tp + fp, 1)
    rec = tp / max(tp + fn, 1)
    f1 = 2 * prec * rec / max(prec + rec, 1e-9)
    
    # Business metrics
    gain = business_gain(tp, fp, retained_value, offer_cost)
    roi_val = roi(tp, fp, retained_value, offer_cost)
    
    # Store results
    result = {
        'threshold': float(threshold),
        'precision': float(prec),
        'recall': float(rec),
        'f1': float(f1),
        'gain': int(gain),
        'roi': float(roi_val),
        'tp': int(tp),
        'fp': int(fp),
        'fn': int(fn),
        'tn': int(tn)
    }
    threshold_results.append(result)
    
    # Track best ROI
    if roi_val > best_roi_info['roi']:
        best_roi_info = {
            'threshold': float(threshold),
            'roi': float(roi_val),
            'gain': int(gain),
            'precision': float(prec),
            'recall': float(rec),
            'f1': float(f1)
        }
    
    # Track high recall threshold (≥0.80)
    if rec >= 0.80 and high_recall_threshold is None:
        high_recall_threshold = float(threshold)

# Convert to DataFrame and sort by ROI
threshold_df = pd.DataFrame(threshold_results)
threshold_df_sorted = threshold_df.sort_values('roi', ascending=False)

print(f"\n🏆 Best ROI threshold: {best_roi_info['threshold']:.3f}")
print(f"   ROI: {best_roi_info['roi']:.3f}")
print(f"   Gain: Rp {best_roi_info['gain']:,}")
print(f"   Precision: {best_roi_info['precision']:.3f}")
print(f"   Recall: {best_roi_info['recall']:.3f}")

if high_recall_threshold:
    print(f"\n📈 High recall threshold (≥0.80): {high_recall_threshold:.3f}")
else:
    print(f"\n📈 No threshold achieved recall ≥0.80")

# Save top 5 thresholds by ROI
top5_roi = threshold_df_sorted.head(5)[['threshold', 'precision', 'recall', 'f1', 'gain', 'roi']]
top5_roi.to_csv('artifacts/threshold_roi_top5.csv', index=False)

print(f"\n📊 Top 5 Thresholds by ROI:")
print(top5_roi.round(4))

print(f"\n💾 Saved top 5 ROI thresholds to artifacts/threshold_roi_top5.csv")

# Store best threshold for later use
thr_best_roi = best_roi_info['threshold']
````

````python
# 13) Eksperimen Perbandingan: Binary vs Churn-Only
print("🔬 EKSPERIMEN PERBANDINGAN: BINARY vs CHURN-ONLY TRAINING")
print("=" * 65)

# 13.1) Setup Split Tambahan untuk Validasi Threshold
print("🔀 Setting up additional validation split for fair threshold tuning...")

# Split training data into train/validation for threshold tuning
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train, 
    test_size=0.2, 
    stratify=y_train, 
    random_state=RANDOM_STATE
)

print(f"✅ Additional split completed:")
print(f"   Training: {X_tr.shape[0]} samples")
print(f"   Validation: {X_val.shape[0]} samples") 
print(f"   Test (holdout): {X_test.shape[0]} samples")
print(f"   Train churn rate: {y_tr.mean():.3f}")
print(f"   Validation churn rate: {y_val.mean():.3f}")

# Helper function for ROI-based threshold selection
def pick_best_threshold_by_roi(y_true, scores, retained_value=500_000, offer_cost=50_000):
    """Find best threshold based on ROI optimization"""
    best = {"thr": None, "roi": -1e9, "gain": -1e9, "prec": None, "rec": None, "f1": None}
    
    for thr in np.linspace(0.01, 0.99, 99):
        y_pred = (scores >= thr).astype(int)
        tp = ((y_true==1) & (y_pred==1)).sum()
        fp = ((y_true==0) & (y_pred==1)).sum()
        fn = ((y_true==1) & (y_pred==0)).sum()
        tn = ((y_true==0) & (y_pred==0)).sum()
        
        g = business_gain(tp, fp, retained_value, offer_cost)
        r = roi(tp, fp, retained_value, offer_cost)
        prec = tp / max(tp+fp, 1)
        rec = tp / max(tp+fn, 1)
        f1 = 2*prec*rec / max(prec+rec, 1e-9)
        
        if r > best["roi"]:
            best.update({
                "thr": float(thr), "roi": float(r), "gain": int(g),
                "prec": float(prec), "rec": float(rec), "f1": float(f1)
            })
    
    return best

print(f"\n✅ ROI threshold optimization function ready")
````

````python
# ==== SAFE BOOTSTRAP + PROBA HELPER ====
import numpy as np
from sklearn.preprocessing import MinMaxScaler

def proba_safe(model, X):
    """Return probability-like scores in [0,1] no matter the estimator."""
    # 1) predict_proba if available
    if hasattr(model, "predict_proba"):
        p = model.predict_proba(X)
        return p[:, 1] if p.ndim == 2 else p
    # 2) decision_function if available (SVM, linear models)
    if hasattr(model, "decision_function"):
        s = model.decision_function(X)
    else:
        # 3) fallback: use raw predict as score
        s = model.predict(X)
    s = np.asarray(s).reshape(-1, 1)
    return MinMaxScaler().fit_transform(s).ravel()

# Pastikan binary_pipeline ada & ter-fit
try:
    binary_pipeline
except NameError:
    binary_pipeline = None

if binary_pipeline is None:
    try:
        # butuh best_pipeline hasil tuning sebelumnya
        binary_pipeline = best_pipeline
    except NameError:
        raise RuntimeError(
            "best_pipeline belum tersedia. Jalankan sel tuning/hyperparameter terlebih dahulu."
        )

# Fit ulang kalau belum di-fit
if not hasattr(binary_pipeline, "classes_"):   # heuristik: belum fit
    binary_pipeline.fit(X_tr, y_tr)

print("✅ binary_pipeline siap dipakai")
````

````python
# 13.3) Rezim B — Churn-Only Training (One-Class) + Fixed Thresholds 0.3/0.5/0.9
print("🎯 REZIM B: CHURN-ONLY TRAINING (ONE-CLASS)")
print("-" * 58)

from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import average_precision_score, roc_auc_score
import numpy as np, json, os

# ----------------- Utility -----------------
def to_01(scores):
    """Min-max normalize to [0,1] for stable fixed-thresholding."""
    s = np.asarray(scores).astype(float)
    lo, hi = np.min(s), np.max(s)
    if hi - lo < 1e-12:
        return np.zeros_like(s)  # all same
    return (s - lo) / (hi - lo)

def evaluate_at_threshold(y_true, scores_01, thr, rv, c, eps=1e-9):
    """Return metrics dict at a fixed threshold (scores already 0..1)."""
    y_pred = (scores_01 >= thr).astype(int)
    tp = int(((y_true == 1) & (y_pred == 1)).sum())
    fp = int(((y_true == 0) & (y_pred == 1)).sum())
    fn = int(((y_true == 1) & (y_pred == 0)).sum())
    tn = int(((y_true == 0) & (y_pred == 0)).sum())
    precision = tp / max(tp + fp, 1)
    recall    = tp / max(tp + fn, 1)
    f1        = 2 * precision * recall / max(precision + recall, eps)
    gain_val  = int(business_gain(tp, fp, rv, c))
    roi_val   = float(roi(tp, fp, rv, c))
    return {
        'precision': float(precision),
        'recall': float(recall),
        'f1': float(f1),
        'gain': gain_val,
        'roi': roi_val,
        'confusion_matrix': {'tp': tp, 'fp': fp, 'fn': fn, 'tn': tn}
    }

# ----------------- 1) Churn-only training data -----------------
X_tr_pos = X_tr[y_tr == 1]
print(f"📊 Churn-only training data: {len(X_tr_pos)} samples (all churn=1)")

# ----------------- 2) Preprocess fit on churn-only -----------------
preprocessor_churn = ColumnTransformer([
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]), numeric_features),
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ]), categorical_features)
])

X_tr_pos_processed = preprocessor_churn.fit_transform(X_tr_pos)
X_val_processed    = preprocessor_churn.transform(X_val)
X_test_processed   = preprocessor_churn.transform(X_test)

print(f"✅ Preprocessor fitted on churn-only data")
print(f"   Processed feature shape: {X_tr_pos_processed.shape[1]}")

# ----------------- 3) Train candidate one-class models -----------------
print(f"\n🤖 Training one-class models...")

svm_params = [
    {'nu': 0.05, 'gamma': 'scale'},
    {'nu': 0.10, 'gamma': 'scale'},
    {'nu': 0.20, 'gamma': 'scale'},
    {'nu': 0.10, 'gamma': 'auto'},
]
if_params = [
    {'contamination': 0.05},
    {'contamination': 0.10},
    {'contamination': 0.20},
]

best_oneclass_model = None
best_oneclass_roi   = {'roi': -1e18}
best_oneclass_type  = None
best_oneclass_thr   = None

def pick_best_threshold_by_roi_SAFE(y_true, raw_scores, rv, c):
    """
    Wrapper: normalize raw_scores to 0..1 before sweeping threshold for ROI.
    Assumes larger score => lebih 'churn-like'.
    """
    s01 = to_01(raw_scores)
    # gunakan seluruh nilai unik sebagai kandidat
    candidates = np.unique(np.concatenate(([0.0], s01, [1.0])))
    best = {'roi': -1e18, 'gain': -1e18, 'thr': 0.5}
    for thr in candidates:
        res = evaluate_at_threshold(y_true, s01, thr, rv, c)
        # tie-breaker: ROI >, lalu Gain >, lalu Recall >
        if (res['roi'] > best['roi'] + 1e-9) or \
           (abs(res['roi'] - best['roi']) <= 1e-9 and res['gain'] > best['gain']) or \
           (abs(res['roi'] - best['roi']) <= 1e-9 and res['gain'] == best['gain'] and res['recall'] > evaluate_at_threshold(y_true, s01, best['thr'], rv, c)['recall']):
            best = {'roi': res['roi'], 'gain': res['gain'], 'thr': float(thr)}
    return best

# Try OneClassSVM
print(f"🔍 Testing OneClassSVM variants...")
for params in svm_params:
    model = OneClassSVM(kernel='rbf', **params)
    model.fit(X_tr_pos_processed)
    val_scores_raw = model.decision_function(X_val_processed)  # raw, not 0..1
    roi_result = pick_best_threshold_by_roi_SAFE(y_val, val_scores_raw, retained_value, offer_cost)
    if roi_result['roi'] > best_oneclass_roi['roi']:
        best_oneclass_model = model
        best_oneclass_roi   = roi_result
        best_oneclass_type  = f"OneClassSVM(nu={params['nu']}, gamma='{params['gamma']}')"
        best_oneclass_thr   = roi_result['thr']
print(f"   Best SVM ROI: {best_oneclass_roi['roi']:.4f}")

# Try IsolationForest
print(f"🔍 Testing IsolationForest variants...")
for params in if_params:
    model = IsolationForest(n_estimators=400, random_state=RANDOM_STATE, **params)
    model.fit(X_tr_pos_processed)
    val_scores_raw = model.score_samples(X_val_processed)  # raw, not 0..1
    roi_result = pick_best_threshold_by_roi_SAFE(y_val, val_scores_raw, retained_value, offer_cost)
    if roi_result['roi'] > best_oneclass_roi['roi']:
        best_oneclass_model = model
        best_oneclass_roi   = roi_result
        best_oneclass_type  = f"IsolationForest(contamination={params['contamination']})"
        best_oneclass_thr   = roi_result['thr']
print(f"   Best IF ROI: {best_oneclass_roi['roi']:.4f}")

print(f"\n🏆 Best one-class model: {best_oneclass_type}")
print(f"   Validation ROI: {best_oneclass_roi['roi']:.4f}")
print(f"   Optimal threshold (on normalized score): {best_oneclass_thr:.4f}")

# ----------------- 4) Final eval on TEST (best ROI threshold) -----------------
print(f"\n📊 Churn-only regime evaluation on test set (best ROI thr):")

# Get raw test scores & normalize to 0..1
if 'OneClassSVM' in best_oneclass_type:
    test_scores_raw = best_oneclass_model.decision_function(X_test_processed)
else:
    test_scores_raw = best_oneclass_model.score_samples(X_test_processed)

test_scores_01 = to_01(test_scores_raw)
y_test_pred_oneclass = (test_scores_01 >= best_oneclass_thr).astype(int)

tp_oc = int(((y_test == 1) & (y_test_pred_oneclass == 1)).sum())
fp_oc = int(((y_test == 0) & (y_test_pred_oneclass == 1)).sum())
fn_oc = int(((y_test == 1) & (y_test_pred_oneclass == 0)).sum())
tn_oc = int(((y_test == 0) & (y_test_pred_oneclass == 0)).sum())

precision_oc = tp_oc / max(tp_oc + fp_oc, 1)
recall_oc    = tp_oc / max(tp_oc + fn_oc, 1)
f1_oc        = 2 * precision_oc * recall_oc / max(precision_oc + recall_oc, 1e-9)

oneclass_results = {
    'regime': 'churn-only',
    'model': best_oneclass_type,
    'threshold_best_roi': float(best_oneclass_thr),
    'pr_auc': float(average_precision_score(y_test, test_scores_01)),
    'roc_auc': float(roc_auc_score(y_test, test_scores_01)),
    'precision': float(precision_oc),
    'recall': float(recall_oc),
    'f1': float(f1_oc),
    'gain': int(business_gain(tp_oc, fp_oc, retained_value, offer_cost)),
    'roi': float(roi(tp_oc, fp_oc, retained_value, offer_cost)),
    'confusion_matrix': {'tp': tp_oc, 'fp': fp_oc, 'fn': fn_oc, 'tn': tn_oc}
}

print(f"   PR AUC: {oneclass_results['pr_auc']:.4f}")
print(f"   ROC AUC: {oneclass_results['roc_auc']:.4f}")
print(f"   Precision: {oneclass_results['precision']:.4f}")
print(f"   Recall: {oneclass_results['recall']:.4f}")
print(f"   F1: {oneclass_results['f1']:.4f}")
print(f"   Gain: Rp {oneclass_results['gain']:,}")
print(f"   ROI: {oneclass_results['roi']:.4f}")

# ----------------- 5) ALSO evaluate fixed thresholds: 0.3 / 0.5 / 0.9 -----------------
print("\n🧪 Fixed-threshold comparison on TEST (0.3, 0.5, 0.9):")
fixed_thresholds = [0.3, 0.5, 0.9]
fixed_results = {}
for thr in fixed_thresholds:
    res = evaluate_at_threshold(y_test, test_scores_01, thr, retained_value, offer_cost)
    fixed_results[f"{thr:.1f}"] = res
    print(f"\n--- Threshold = {thr:.1f} ---")
    print(f"Precision: {res['precision']:.4f}")
    print(f"Recall: {res['recall']:.4f}")
    print(f"F1: {res['f1']:.4f}")
    print(f"Gain: Rp {res['gain']:,}")
    print(f"ROI: {res['roi']:.4f}")
    cm = res['confusion_matrix']
    print(f"Confusion Matrix: TP={cm['tp']}, FP={cm['fp']}, FN={cm['fn']}, TN={cm['tn']}")

# ----------------- 6) Save -----------------
os.makedirs('artifacts', exist_ok=True)
with open('artifacts/compare_oneclass.json', 'w') as f:
    json.dump(oneclass_results, f, indent=2)
with open('artifacts/compare_oneclass_fixed.json', 'w') as f:
    json.dump(fixed_results, f, indent=2)

print("\n💾 Saved:")
print("   - artifacts/compare_oneclass.json (best ROI thr)")
print("   - artifacts/compare_oneclass_fixed.json (fixed thr 0.3/0.5/0.9)")

print("\n⚠️  Note: Feature importance not available for one-class models")
````

# 🎯 Optimasi Threshold untuk RECALL TINGGI — Business Parameters

Bagian ini mengimplementasikan **strategi baru** yang memprioritaskan **RECALL TINGGI** untuk menangkap sebanyak mungkin nasabah yang akan churn, sambil tetap menjaga profitabilitas:

## 🔄 **Perubahan Strategi Optimasi:**
- **SEBELUM**: Maksimalkan ROI → Threshold tinggi (0.99) → Precision tinggi tapi recall rendah
- **SEKARANG**: Maksimalkan RECALL → Threshold rendah → Tangkap lebih banyak churners

## 💰 **Business Logic:**
- **RETAINED_VALUE**: Rp 500,000 (nilai nasabah yang diselamatkan)
- **OFFER_COST**: Rp 50,000 (biaya kampanye per nasabah)
- **Rasio Cost**: 10:1 - kehilangan churner 10x lebih mahal dari false alarm
- **ROI Minimal**: 1.5x (tetap profitable)

## 🎯 **Proses Optimasi Baru:**
1. **Sweep semua threshold** dari 0.01 sampai 0.99
2. **Filter threshold** yang masih profitable (ROI ≥ 1.5x)
3. **Pilih threshold** dengan **RECALL TERTINGGI** dari yang profitable
4. **Evaluasi trade-off** antara recall vs precision
5. **Bandingkan** dengan strategi ROI lama

## 📊 **Expected Impact:**
- **Recall Target**: 75-85% (vs 56% sebelumnya)
- **Missing Churners**: Turun drastis dari 43% ke ~20%
- **ROI**: Masih profitable 2-4x (vs 8.9x sebelumnya)
- **Strategy**: Lebih cocok untuk bank yang prioritas retensi nasabah

````python
# ============================================================
# KONSTANTA BISNIS & FUNGSI OPTIMASI RECALL (idempotent)
# ============================================================
try:
    RETAINED_VALUE
    OFFER_COST
except NameError:
    RETAINED_VALUE = 500_000  # nilai yang diselamatkan per nasabah retained (Rp)
    OFFER_COST     = 50_000   # biaya kampanye per nasabah ditarget (Rp)

import numpy as np
import pandas as pd
from sklearn.metrics import average_precision_score, roc_auc_score

def business_gain(tp, fp, retained_value=RETAINED_VALUE, offer_cost=OFFER_COST):
    """Gain = Revenue - Cost."""
    revenue = tp * retained_value
    cost = (tp + fp) * offer_cost
    return revenue - cost

def roi(tp, fp, retained_value=RETAINED_VALUE, offer_cost=OFFER_COST):
    """ROI = (Revenue - Cost) / Cost."""
    cost = (tp + fp) * offer_cost
    return np.inf if cost <= 0 else (business_gain(tp, fp, retained_value, offer_cost) / cost)

def recall_score_custom(tp, fn):
    """Hitung recall = TP / (TP + FN)"""
    return tp / max(tp + fn, 1)

def precision_score_custom(tp, fp):
    """Hitung precision = TP / (TP + FP)"""
    return tp / max(tp + fp, 1)

def _confusion_counts(y_true, y_pred):
    tp = int(((y_true == 1) & (y_pred == 1)).sum())
    fp = int(((y_true == 0) & (y_pred == 1)).sum())
    fn = int(((y_true == 1) & (y_pred == 0)).sum())
    tn = int(((y_true == 0) & (y_pred == 0)).sum())
    return tp, fp, fn, tn

def evaluate_at_threshold(y_true, scores, thr,
                          retained_value=RETAINED_VALUE, offer_cost=OFFER_COST):
    """Hitung metrik + ROI pada threshold tertentu."""
    y_pred = (scores >= thr).astype(int)
    tp, fp, fn, tn = _confusion_counts(y_true, y_pred)
    prec = precision_score_custom(tp, fp)
    rec  = recall_score_custom(tp, fn)
    f1   = 0.0 if (prec + rec) == 0 else 2 * prec * rec / max(prec + rec, 1e-9)
    gain = business_gain(tp, fp, retained_value, offer_cost)
    r    = roi(tp, fp, retained_value, offer_cost)
    try:    pr_auc = average_precision_score(y_true, scores)
    except: pr_auc = np.nan
    try:    rocauc = roc_auc_score(y_true, scores)
    except: rocauc = np.nan
    return {
        "threshold": float(thr),
        "tp": tp, "fp": fp, "fn": fn, "tn": tn,
        "precision": float(prec), "recall": float(rec), "f1": float(f1),
        "gain": int(gain), "roi": float(r),
        "pr_auc": float(pr_auc), "roc_auc": float(rocauc)
    }

def sweep_thresholds(y_true, scores, step=0.01,
                     retained_value=RETAINED_VALUE, offer_cost=OFFER_COST):
    """Sweep dari 0.01 sampai 0.99 dan hitung semuanya."""
    return pd.DataFrame([
        evaluate_at_threshold(y_true, scores, thr, retained_value, offer_cost)
        for thr in np.arange(step, 1.0, step)
    ])

def pick_best_threshold_by_recall(y_true, scores, min_roi=1.0, step=0.01,
                                  retained_value=RETAINED_VALUE, offer_cost=OFFER_COST):
    """Ambil threshold dengan RECALL tertinggi yang masih profitable (ROI >= min_roi)."""
    df = sweep_thresholds(y_true, scores, step, retained_value, offer_cost)
    
    # Filter hanya threshold yang profitable (ROI >= min_roi)
    profitable_df = df[df['roi'] >= min_roi]
    
    if len(profitable_df) == 0:
        print(f"⚠️ Tidak ada threshold dengan ROI >= {min_roi}, menggunakan ROI tertinggi")
        profitable_df = df.sort_values("roi", ascending=False).head(1)
    
    # Dari yang profitable, pilih yang recall tertinggi
    best_recall = profitable_df.sort_values("recall", ascending=False).head(5)
    
    return best_recall.iloc[0].to_dict(), best_recall

# Legacy function untuk compatibility
def pick_best_threshold_by_roi(y_true, scores, step=0.01,
                               retained_value=RETAINED_VALUE, offer_cost=OFFER_COST):
    """Ambil threshold dengan ROI tertinggi + 5 Teratas untuk laporan."""
    df = sweep_thresholds(y_true, scores, step, retained_value, offer_cost)
    top5 = df.sort_values("roi", ascending=False).head(5)
    return top5.iloc[0].to_dict(), top5

print("✅ Parameter bisnis dan fungsi OPTIMASI RECALL berhasil didefinisikan")
print(f"💰 RETAINED_VALUE: Rp {RETAINED_VALUE:,}")
print(f"💸 OFFER_COST: Rp {OFFER_COST:,}")
print(f"🎯 STRATEGI: Maksimalkan RECALL dengan ROI minimal 1.0x")
print(f"📊 Kalkulasi siap untuk optimasi threshold fokus recall")
````

````python
# ============================================================
# SELEKSI RECALL (Validation) -> EVALUASI (Test) + ARTEFAK
# ============================================================
from pathlib import Path

def _safe_predict_scores(model, X):
    if hasattr(model, "predict_proba"):
        return model.predict_proba(X)[:, 1]
    if hasattr(model, "decision_function"):
        return model.decision_function(X)
    raise RuntimeError("Model tidak punya predict_proba/decision_function")

try:
    # Pastikan variabel ini sudah ada dari sel sebelumnya:
    # best_pipeline, X_val, y_val, X_test, y_test
    _ = best_pipeline; X_val; y_val; X_test; y_test

    Path("artifacts").mkdir(parents=True, exist_ok=True)

    # --- BINARY - OPTIMASI UNTUK RECALL TINGGI ---
    y_val_scores  = _safe_predict_scores(best_pipeline, X_val)
    y_test_scores = _safe_predict_scores(best_pipeline, X_test)

    # STRATEGI BARU: Maksimalkan recall dengan ROI minimal 1.5x
    best_thr_info, top5_recall = pick_best_threshold_by_recall(
        y_val, y_val_scores, min_roi=1.5, step=0.01,
        retained_value=RETAINED_VALUE, offer_cost=OFFER_COST
    )
    print(">> THRESHOLD TERBAIK (VAL) berdasarkan RECALL MAKSIMAL (ROI >= 1.5x) — BINARY:")
    print(best_thr_info)
    print(f"🎯 PRIORITAS: Recall = {best_thr_info['recall']:.1%} | Precision = {best_thr_info['precision']:.1%}")
    print(f"💰 ROI = {best_thr_info['roi']:.2f}x (masih profitable)")

    # Simpan top 5 berdasarkan recall (bukan ROI)
    top5_recall.to_csv("artifacts/threshold_recall_top5.csv", index=False)

    thr = best_thr_info["threshold"]
    test_eval = evaluate_at_threshold(
        y_test, y_test_scores, thr,
        retained_value=RETAINED_VALUE, offer_cost=OFFER_COST
    )
    print("\n>> EVALUASI TEST @ threshold RECALL terbaik — BINARY")
    print(f"🎯 Recall: {test_eval['recall']:.1%} (PRIORITAS UTAMA)")
    print(f"⚖️ Precision: {test_eval['precision']:.1%}")
    print(f"💰 ROI: {test_eval['roi']:.2f}x")
    print(f"💵 Keuntungan: Rp {test_eval['gain']:,}")

    # Analisis trade-off
    missing_churners = test_eval['fn']
    false_alarms = test_eval['fp']
    total_churners = test_eval['tp'] + test_eval['fn']
    
    print(f"\n📊 ANALISIS BUSINESS IMPACT:")
    print(f"   ✅ Churners Tertangkap: {test_eval['tp']} dari {total_churners} ({test_eval['recall']:.1%})")
    print(f"   😱 Churners Terlewat: {missing_churners} ({missing_churners/total_churners:.1%})")
    print(f"   ⚠️ False Alarms: {false_alarms} nasabah")
    print(f"   💸 Cost False Alarms: Rp {false_alarms * OFFER_COST:,}")
    print(f"   💔 Cost Missing Churners: Rp {missing_churners * RETAINED_VALUE:,}")

    # Simpan config serving dengan fokus recall
    import json
    with open("artifacts/serving_config_recall_focused.json", "w", encoding="utf-8") as f:
        json.dump({
            "mode": "binary_recall_focused",
            "strategy": "maximize_recall_with_profitable_roi",
            "threshold": thr,
            "min_roi_required": 1.5,
            "performance": {
                "recall": test_eval["recall"],
                "precision": test_eval["precision"],
                "roi": test_eval["roi"],
                "f1": test_eval["f1"]
            },
            "retained_value": RETAINED_VALUE,
            "offer_cost": OFFER_COST,
            "business_rationale": "Bank prioritizes catching all potential churners over campaign precision"
        }, f, ensure_ascii=False, indent=2)

    # --- PERBANDINGAN: ROI-focused vs Recall-focused ---
    print(f"\n" + "="*60)
    print("📊 PERBANDINGAN: STRATEGI ROI vs RECALL")
    print("="*60)
    
    # Hitung juga untuk perbandingan dengan strategi ROI lama
    roi_thr_info, top5_roi = pick_best_threshold_by_roi(
        y_val, y_val_scores, step=0.01,
        retained_value=RETAINED_VALUE, offer_cost=OFFER_COST
    )
    
    roi_test_eval = evaluate_at_threshold(
        y_test, y_test_scores, roi_thr_info["threshold"],
        retained_value=RETAINED_VALUE, offer_cost=OFFER_COST
    )
    
    comparison_data = {
        "Strategy": ["ROI-Focused", "RECALL-Focused"],
        "Threshold": [roi_thr_info["threshold"], best_thr_info["threshold"]],
        "Recall": [roi_test_eval["recall"], test_eval["recall"]],
        "Precision": [roi_test_eval["precision"], test_eval["precision"]],
        "ROI": [roi_test_eval["roi"], test_eval["roi"]],
        "Missing_Churners": [roi_test_eval["fn"], test_eval["fn"]],
        "False_Alarms": [roi_test_eval["fp"], test_eval["fp"]]
    }
    
    comparison_df = pd.DataFrame(comparison_data)
    print(comparison_df.round(3))
    
    # Highlight the wins
    recall_improvement = (test_eval["recall"] - roi_test_eval["recall"]) * 100
    churners_saved = roi_test_eval["fn"] - test_eval["fn"]
    
    print(f"\n🎯 KEUNTUNGAN STRATEGI RECALL:")
    print(f"   📈 Peningkatan Recall: +{recall_improvement:.1f} percentage points")
    print(f"   💝 Churners Tambahan Tertangkap: {churners_saved} nasabah")
    print(f"   💰 Nilai Nasabah Diselamatkan: Rp {churners_saved * RETAINED_VALUE:,}")
    print(f"   ⚠️ Trade-off: ROI turun dari {roi_test_eval['roi']:.2f}x ke {test_eval['roi']:.2f}x")

    # --- OPSIONAL: ONE-CLASS (kalau tersedia) ---
    compare_rows = [{
        "regime": "binary_recall_focused",
        "threshold": thr,
        "precision": test_eval["precision"],
        "recall": test_eval["recall"],
        "f1": test_eval["f1"],
        "gain": test_eval["gain"],
        "roi": test_eval["roi"]
    }]

    # Jika ada best_oneclass_model & preprocessor_churn (hasil preprocessing), jalankan juga
    try:
        _ = best_oneclass_model; _ = preprocessor_churn  # pastikan variabel ini tersedia dari notebook
        scores_val_oc  = best_oneclass_model.decision_function(preprocessor_churn.transform(X_val))
        scores_test_oc = best_oneclass_model.decision_function(preprocessor_churn.transform(X_test))

        best_thr_info_oc, top5_oc = pick_best_threshold_by_recall(
            y_val, scores_val_oc, min_roi=1.0, step=0.01,
            retained_value=RETAINED_VALUE, offer_cost=OFFER_COST
        )
        print("\n>> THRESHOLD TERBAIK (VAL) berdasarkan RECALL — ONE-CLASS:")
        print(best_thr_info_oc)

        thr_oc = best_thr_info_oc["threshold"]
        test_oc = evaluate_at_threshold(
            y_test, scores_test_oc, thr_oc,
            retained_value=RETAINED_VALUE, offer_cost=OFFER_COST
        )
        print("\n>> EVALUASI TEST @ threshold RECALL terbaik — ONE-CLASS")
        print(test_oc)

        compare_rows.append({
            "regime": "one-class_recall_focused",
            "threshold": thr_oc,
            "precision": test_oc["precision"],
            "recall": test_oc["recall"],
            "f1": test_oc["f1"],
            "gain": test_oc["gain"],
            "roi": test_oc["roi"]
        })

    except Exception as e:
        print("\nINFO: Model one-class tidak tersedia atau variabelnya berbeda. Lewati perbandingan.")
        print("Detail:", repr(e))

    # Simpan tabel perbandingan (fokus recall)
    pd.DataFrame(compare_rows).to_csv("artifacts/compare_recall_focused.csv", index=False)
    print(f"\n💾 Tersimpan:")
    print(f"   📊 artifacts/threshold_recall_top5.csv")
    print(f"   ⚙️ artifacts/serving_config_recall_focused.json")
    print(f"   📋 artifacts/compare_recall_focused.csv")

except Exception as e:
    print("INFO: Jalankan sel ini setelah variabel tersedia:")
    print("- best_pipeline")
    print("- X_val, y_val")
    print("- X_test, y_test")
    print("Detail:", repr(e))
````
